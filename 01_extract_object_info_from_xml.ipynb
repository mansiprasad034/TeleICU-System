```python
import glob
import xml.etree.ElementTree as et

# Load all xml files and store in a list
xml_list = glob.glob('./data_images/*.xml')

# Data cleaning
# Replace \\ with /
xml_files = list(map(lambda x: x.replace('\\', '/'), xml_list))

# Step 2: read xml file
# From each xml we need to extract
# filename, size(width, height), object(name, xmin, ymin, xmax and ymax)
def extract_text(filename):
    tree = et.parse(filename)
    root = tree.getroot()

    # Extract filename
    image_name = root.find('filename').text
    print(image_name)

    # Width and height of the image
    width = root.find('size').find('width').text
    print(width)

    # Height
    height = root.find('size').find('height').text
    print(height)

    parser = []
    for obj in root.findall('object'):
        name = obj.find('name').text
        bndbox = obj.find('bndbox')
        xmin = bndbox.find('xmin').text
        xmax = bndbox.find('xmax').text
        ymin = bndbox.find('ymin').text
        ymax = bndbox.find('ymax').text
        parser.append([image_name, width, height, name, xmin, xmax, ymin, ymax])
    return parser

# Extract and print data from all xml files
parser_all = []
for xml_file in xml_files:
    parser_all.extend(extract_text(xml_file))

print(parser_all)
```

```python
import os
from shutil import move

def save_data(filename, folder_path, group_obj):
    # Move image
    src = os.path.join('data_images', filename)  # Corrected typo here
    dst = os.path.join(folder_path, filename)
    
    # Debug: Print source and destination paths
    print(f"Source path: {src}")
    print(f"Destination path: {dst}")
    
    try:
        move(src, dst)
    except FileNotFoundError as e:
        print(f"Error: {e}")
        return
    
    # Save the labels
    text_filename = os.path.join(folder_path, os.path.splitext(filename)[0] + '.txt')
    group_obj.get_group(filename).set_index('filename').to_csv(
        text_filename, sep=' ', index=False, header=False
    )

# Assuming train_folder and groupby_obj_train are already defined
# Debug: Print the first few elements of filename_series
print(filename_series.head(2))

# Apply the save_data function
filename_series.head(2).apply(save_data, args=(train_folder, groupby_obj_train))
```
